LONGEST DURATION MOVIE IS MADE BY EGYPTIANS ACCORDING TO NETFLIX BEFORE SEPTEMBER 9,2020

```python
longest_duration=netflix_df["duration"].max()
longest_duration_movie_title = netflix_df[netflix_df['duration'] == longest_duration]['title'].values[0]
print(longest_duration_movie_title)
```

```python
longest_duration_movie_country = netflix_df[netflix_df['duration'] == longest_duration]['country'].values[0]
print(longest_duration_movie_country)
```

UNITED STATES WON HERE TOO WITH A MARGIN OF DOUBLE

```python
netflix_df['country'].value_counts()
```

```python
country_counts = netflix_df['country'].value_counts()
plt.figure(figsize=(25, 10))
country_counts.plot(kind='bar', color='skyblue')
plt.title('Count of Movies for Each Country')
plt.xlabel('Country')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
