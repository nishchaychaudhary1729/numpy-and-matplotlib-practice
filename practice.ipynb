```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
v0 = 20  # Initial velocity (m/s)
theta = np.radians(45)  # Launch angle in radians
g = 9.81  # Acceleration due to gravity (m/s^2)

# Calculate initial velocities in x and y directions
v0x = v0 * np.cos(theta)
v0y = v0 * np.sin(theta)

# Time array
t_max = (2 * v0y) / g  # Time of flight
t = np.linspace(0, t_max, 100)

# Calculate x and y coordinates
x = v0x * t
y = v0y * t - 0.5 * g * t**2

# Plot trajectory
plt.plot(x, y)

# Add labels and title
plt.xlabel('Horizontal Distance (m)')
plt.ylabel('Vertical Distance (m)')
plt.title('Projectile Motion')

# Show plot
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

g=9.8
v=20
theta=np.radians(45)
v0y=2*v*np.sin(theta)
t_max=(2*v0y)/g
t=np.linspace(0,t_max,100)
v0x=2*v*np.cos(theta)
x=v0x*t
y=v0y*t-0.5*g*(t**2)
plt.plot(x, y)
plt.grid()
plt.xlabel("distance(in m)")
plt.ylabel("height(in m)")
plt.title('projectile motion')
plt.axvline(x[len(x)//2],ymax=0.95,alpha=0.4)
plt.text(x[len(x)//2],max(y)+2,"maximum height",fontsize="10",horizontalalignment='center',verticalalignment='top')
plt.show()
```

```python
r=50
l=0.1
c=0.001
vmax=100
f=50
w=2*np.pi*f
xl=w*l
xc=1/(w*c)
theta=np.arctan2(xl,r)
time = np.arange(0,0.1,0.0001)
cur_r=vmax*np.cos(w*time)/r
cur_l=vmax*np.cos(w*time+theta)/xl
plt.plot(time,cur_l)
plt.plot(time,cur_r)
plt.show()
```
